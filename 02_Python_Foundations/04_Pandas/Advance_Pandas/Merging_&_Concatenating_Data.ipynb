# Merging & Concatenating Data

```python
import pandas as pd
```

```python
order = pd.read_csv(r"D:\AI ML ENGINEERING JOURNY\02_Python_Foundations\04_Pandas\dataset\orders.csv")
```

```python
product = pd.read_csv(r"D:\AI ML ENGINEERING JOURNY\02_Python_Foundations\04_Pandas\dataset\products.csv")
```

 ## import both order and product dataset

```python
merged_df = pd.merge(order, product, on="product_id", how="inner")
```

```python
merged_df.head()

```

```python
merged_df.shape
```

```python
merged_df["total_value"] = merged_df["quantity"] * merged_df["unit_price"]
```

```python
merged_df
```

```python
merged_df.groupby("category")["total_value"].sum()
```

```python
merged_df.groupby("brand")["total_value"].sum()
```
