# Handling Null Values in the `age` Column

### Code Used

```python
students_performance["age"] = students_performance["age"].fillna(
    students_performance["age"].mode()[0]
)
```

---

## What This Code Does (Step-by-Step)

### 1️⃣ Select the Column
```python
students_performance["age"]
```
Selects the `age` column from the DataFrame.

---

### 2️⃣ Fill Missing Values
```python
.fillna(students_performance["age"].mode()[0])
```

- Finds the **mode** (most frequent value) of the `age` column.
- Replaces only the `NaN` (missing) values.
- Does NOT modify existing valid values.

---

### 3️⃣ Assign Back to the DataFrame
```python
students_performance["age"] = ...
```

- Updates the original `age` column.
- Ensures changes are saved properly.
- Avoids chained assignment errors.
